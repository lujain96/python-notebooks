## Problem 1

### <span style="color:red">Hangman Game</span>
#### Build a prgram that simulaates hangman game, you can find the game discription [Here](https://en.wikipedia.org/wiki/Hangman_(game)).

#### The game starts with choose the level following the rules in the table below, then takes an input sentence from player_1 and then goes in a loop with player_2 trying to guess the sentence player one insert.

#### Player_2 can guess a number of charcters spicified below, or he also can guess the full sentence at once and this will be the end of the game 

##### Level details

| level | # if total guesses |
|-|-|
| Easy | # of alphabets in the sentence + 10 |
| Meduim| # of alphabets in the sentence + 7 |
|Hard|  # of alphabets in the sentence + 5 |

#### Output Discription:
##### You should have seven standard outputs:
##### Note: All the below casses are studied under the following test case:
- Right answer is: Tahaluf Al Emarat
- Wrong charcters input are: k - s - n - v
|Case|Output|
| -- | -- |
| Wrong Answer| T-h--u- -- Em-r-t  <br> Wrong Chars: k - s - n - v|
|Right Answer|Greate Guess: a <br> Taha-u- A- Emarat |
|Last Trial if wrong|You ran out of trials and Catched T-h--u- -- Em-r-t <br> The right sentence is Tahaluf AL Emarat|
|Full Sentence Right| WOW you are brave, you got it right from the [number of attempts he made] trial|
|Full Sentence Wrong|WOW you are brave, but it was a wrong answer <br> The right one is Tahaluf Al Emarat|
|Not an alphabets|You entered a non-alphabet character|
|Already Choosed Alphabete|You already entered this character and it was [RIGHT/WRONG]|

#### Notes:
- DON'T count the repated alphabets.
- You should handle the spaces between words as SPACE is not considered as a character in our game
- If player_2 inserts more than one charcter, you must handle his input as the full sentence.
- If player_2 inserts any charcter not in the alphabets you should reject it and not cnosider it as a trial.
- If player_2 inserts an alphabet that he choosed before you should notify him and not consider it as a trial.
- The input must be striped from any added spaces or newlines charcters in the start and the end.
- You should delete any non-alphabets characters in the input sentence from player_1.
- You must handle the upeer and lower cases in the player_1 sentnces and player_2 inputs, as both should not worry about them

#### Applied Concepts
- List Methods
- Slicing a list
- Conditional test with lists
- Nested Lists
#### Feel free to code with any of dictionries or lists.


```python
### Write Your Code Here ###


word = input('Enter the word for player 1: ').lower()
word = ''.join([i for i in word if i.isalpha()])
word = word.split(' ')
words = "" 
# traverse in the string  
for ele in word: 
    words += ele 

attempts = sum(len(i) for i in word)
count = 1
wrong_chars = []


while True:
    level = input('Choose your level: Easy, Medium, or Hard? ').strip().lower()
    if level == 'easy' or level == 'medium' or level == 'hard':
        break
    else:
        print("This is not a valid level. Please enter a valid level")
        continue
if level == 'easy':
    attempts+= 10
elif level == 'medium':
    attempts+= 7
else:
    attempts+= 5

print('You have', attempts, 'attempts') 
    
guesses = ''

while attempts > 0:
    
    wrong = 0
    for char in words:
        #compare the character with guesses
        if char in guesses:
            print(char) 
        else:
            print('_')
             
            # wrong will be incremented by 1 each time the user input is not true
            wrong += 1          
            
 
    if wrong == 0:
        # print 'You win' and the correct word if wrong is 0, then end the game 
        print('You Win')
        print('The word is:', words)
        break
    
    
    word_guess = input("guess a character:").strip().lower()
    if word_guess == len(word_guess) * word_guess[0]:
        #check if the input is alphabet or not
        #if not, the user has to re-enter the input and type an alphabet
        if word_guess.isalpha() != True:
            print('You\'ve entered a wrong value, please enter a valid one.')
            continue
        #check if the user entered a character more than once
        #if yes, the user has to re-enter the input and type another alphabet    
        if word_guess in guesses:
            print('You\'ve already entered this character, please enter another one.')
            continue
        else:
            # every character will be stored in guesses list
            guesses += word_guess
         
        # check input with the character in word
            if word_guess not in words:
                attempts -= 1
                count+=1
            # if the character does not match the word
            # then 'Wrong chars' will be displayed as a list in the output
                wrong_chars.append(word_guess)
                print('Wrong chars:', wrong_chars)
         
            #print the number of attempts left
                print("You have", + attempts, 'more guesses')
                if attempts == 0:
                    print('You lose')
                    break    
    else:
         
        # every input character will be stored in guesses
        print('Great Guess:', word_guess)
        guesses += word_guess
     
        # check input with the character in word
        if word_guess != words:
         
            attempts -= 1
            count+= 1
         
            # if the character does not match the word
            # then end the game
            print("WOW you are brave, but it was a wrong answer\nThe right answer is", words)
            break
        else:
            print('WOW you are brave, you got it right from the', count, 'trial')
            break
```
