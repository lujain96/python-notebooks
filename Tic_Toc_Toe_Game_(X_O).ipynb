```python
### Write Your Program Here ###

''' The game board will be a dictionary 
    keys will be from 1 to 9 starting from the bottom left and so on..
    the keys will have empty values initially, and then the values will be filled after each move
    '''

board = {'7': ' ' , '8': ' ' , '9': ' ',
        '4': ' ' , '5': ' ' , '6': ' ' ,
        '1': ' ' , '2': ' ' , '3': ' ' }
keys = []
for key in board:
    keys.append(key)

def printBoard(board):
    ''' A function to print the board after every move
    args: board (dict)
    actions: print the updated board starting from the bottom left
    '''
    print(board['7'] + '|' + board['8'] + '|' + board['9'])
    print('-----')
    print(board['4'] + '|' + board['5'] + '|' + board['6'])
    print('-----')
    print(board['1'] + '|' + board['2'] + '|' + board['3'])
           
    
def game():
    '''
    game() function, where the game starts
    at each step, the turn will be for one of (X, O)
    when the player chooses a place, the function will check if it's empty 
    then it will check if there is a winner after 5 moves for every move.
    '''
    turn = 'X' #initially
    count = 0
    places = ['1','2','3','4','5','6','7','8','9']
    #loop over the game
    for i in range(10):
        printBoard(board)
        print("It's " + turn + "\'s turn")
        move = input('Which place do you want to take?')
        if move in places:
            if board[move] == ' ':
                board[move] = turn
                count += 1
            else:
                print("That place is filledm please choose another one")
                continue
        else:
            print('Wrong value, please enter a valid place')
            continue
            
        # check who's the winner after 5 moves. 
        if count >= 5:
            #check the rows first
            if board['7'] == board['8'] == board['9'] != ' ': # check the top row
                printBoard(board)
                print("Game Over." + turn + " won. ")                  
                break
            elif board['4'] == board['5'] == board['6'] != ' ': # check the middle row
                printBoard(board)
                print("Game Over." + turn + " won. ")    
                break
            elif board['1'] == board['2'] == board['3'] != ' ': # check the bottom row
                printBoard(board)
                print("Game Over." + turn + " won. ")    
                break
            #check the diagonal
            elif board['1'] == board['5'] == board['9'] != ' ': # check diagonal
                printBoard(board)
                print("Game Over." + turn + " won. ")    
                break
            elif board['7'] == board['5'] == board['3'] != ' ': # check diagonal
                printBoard(board)
                print("Game Over." + turn + " won. ")    
                break
            #check the columns
            elif board['1'] == board['4'] == board['7'] != ' ': # check the fisrt column
                printBoard(board)
                print("Game Over." + turn + " won. ")    
                break
            elif board['2'] == board['5'] == board['8'] != ' ': # check the second column
                printBoard(board)
                print("Game Over." + turn + " won. ")  
                break
            elif board['3'] == board['6'] == board['9'] != ' ':  # check the last column
                printBoard(board)
                print("Game Over." + turn + " won. ")
                break
            
        # If all places are filled and no one wins, print 'No one won!'.
        if count == 9:
            print("Game Over. No one won!")
            break

        # change the player at each turn.
        if turn =='X':
            turn = 'O'
        else:
            turn = 'X'        

#start the game
game()
#ask user if they want to play again
while True:
    again = input("Do want to play Again? yes or no ").lower()
    if again == "yes":  
        for key in keys:
            board[key] = " "
        game()
        continue
    elif again == "no":
        break
    else:
        print("Please enter a valid answer ")
        continue
    
```
