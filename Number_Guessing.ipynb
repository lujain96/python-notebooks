## Number Guessing

#### Write a program that generates a random number between 0 - > 100  (help is below), and let the user guess what number you generates, the user has only three quistions to ask following the next format:
- is it greater than {number} 
- is it divisible {number}
- is it divisible {number}

#### These questions should help the user to narrowing his search.
#### There are four special words he can use, with the following meanings:
- exit: he is a loser and tends to exit the game so close your program
- pass: change the number so he can play with another randomly generated number.
- help:  this is the magic word if he says it, you must say if the target number is lower or grater than his guess, each time he guess
- no help: this option close the help option above and returns only true or false for his guess

#### Output Discription
#### Your output have three main paths:
- answering the three quistions with : Yes or No
- responding to his guess with : True or False
- with the help option activated: True or False it's Greater/Lower

#### Notes:
- All numbers must be integres and casted as that.
- all input words must be casted to lower case and handeled like that

#### Applied Concepts
- Control Flow Tools
- If statements
- Loops
- range()
- Break, continue and pass
- Multiple conditions
- Strings Operations

```python
### Write Your Code Here ###
### A program to let the user guess a random number 
import random
random_number = int(random.randint(0, 100)) ### Line to generate random_numbers

#Greeting 
desicion = ''
print('Welcome to the guessing game!\nYou have to guess the number I am hiding, you are allowed to ask me 3 questions')

#The user has 2 questions to ask
#check if they are valid
#question 1
while True:
    question = input('Type \'greater\' if you want to know if the number is greater than a specific number\n').lower()
    if question == 'greater':
        while True:
            try:
                greater_number = int(input('Enter the number : '))
            except ValueError:
                print("This is not a number. Please enter a valid number")
                continue
            else:
                break
        if greater_number < random_number:
            print('yes')
        else:
            print('no') 
        break
    else:
        print("Please enter a valid word")
        continue
        
#question 2
while True:
    question2 = input('Type \' divisible \' if you want to know if it\'s divisible: ').lower()
    if question2 == 'divisible':
        while True:
            try:
                div_number = int(input('Enter the number : '))
            except ValueError:
                print("This is not a number. Please enter a valid number")
                continue
            else:
                break
        if random_number % div_number == 0:
            print('yes')
        else:
            print('no') 
        break
    else:
        print("Please enter a valid word")
        continue
    
#The user now has to guess the number
#check if the number is valid
while (True):
    while True:
        try:
            guess = int(input('Enter your guess now: '))
        except ValueError:
            print("This is not a number. Please enter a valid number")
            continue
        else:
            break
    if guess == random_number: #check if the guess is right
        print('True')
        break
    else:
        print('False')
        
    while True: #The user now has to make a decision, check if it's valid
        decision = input('What\' your decision: exit, pass, help, no help? ').lower()
        if decision == 'exit': #exit the game
            print('Game over.')
            break
        elif decision == 'pass':
            random_number = int(random.randint(0, 100)) #regenerate the number
            continue
        elif decision == 'help': #give a hint
            if guess > random_number:
                print('Your guess is greater')
                break
            elif guess < random_number:
                print('Your guess is lower')
                break
        elif decision == 'no help': #there's nothing to do
            if guess == random_number:
                print('True')
                break
            else:
                print('False') 
            break
        else:
            print('Please enter a valid option.')
            continue

```
